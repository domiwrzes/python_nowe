Ćwiczenie 3: Operacje wektorowe — analiza biznesowa (25
min)

```python
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1: Podstawowe statystyki cenowe

```python
print("Średnia cena: " + str(ceny.mean()))
print("Najtańszy produkt: " + str(ceny.min()))
print("Najdroższy produkt: " + str(ceny.max()))
```

Zadanie 2: Wartość magazynu

```python
wartosc_produkt = ceny * ilosci_magazyn

lacznie_magazyn = sum(wartosc_produkt)

print("Wartości produktów: " + str(wartosc_produkt))
print("Łączna wartość magazynu: " + str(lacznie_magazyn))
```

Zadanie 3: Analiza cenowa
# Stworz array z cenami po rabacie 15% (promocja świąteczna)
# Stworz array z cenami po uwzględnieniu VAT 23%
# Ile produktów kosztuje powyżej 1000 zł?
# Podpowiedź: rabat = ceny * 0.85, VAT = ceny * 1.23, zliczanie: (warunek).sum()

```python
rabat = ceny * 0.85

vat = ceny * 1.23

pow_1000 = sum(ceny > 1000)

print("Ceny po rabacie :" + str(rabat))

print("Ceny z vat: " + str(vat))

print("Ilość produktów powyżej 1000zł: " + str(pow_1000))
```

Zadanie 4: Filtrowanie
# Produkty z oceną >= 4.5 — nazwy i ceny
# Produkty poniżej 200 zł — nazwy i ilości w magazynie
# Podpowiedź: maska = oceny_klientow >= 4.5, potem ceny[maska]
# Uwaga: nazwy to zwykła lista — zamień ją na np.array(nazwy), żeby użyć maski

```python
nazwy_a = np.array(nazwy)
maska = (oceny_klientow >= 4.5)
wys_ocena = nazwy_a[maska]
ceny_ocena = ceny[maska]
print("Produkty z wysoką oceną: " + str(wys_ocena))
print("Ich ceny: " + str(ceny_ocena))

maska_pon = (ceny < 200)
nazwy_pon = nazwy_a[maska_pon]
ilosci_pon = ilosci_magazyn[maska_pon]
print("Produkty poniżej 200zł: " + str(nazwy_pon))
print("Ilości: " + str(ilosci_pon))
```

Zadanie 5: Analiza zapasów
# Produkty z niskim stanem magazynowym (< 30 szt.) — nazwy
# Łączna wartość tych produktów
# Ile % całego magazynu stanowią?
# Podpowiedź: wartosci = ceny * ilosci_magazyn, potem filtruj maską

```python
maska_mal = (ilosci_magazyn < 30)
produkty_mal = nazwy_a[maska_mal]
print("Produkty z niskim stanem magazynowym: " + str(produkty_mal))

wart_mag = (ceny * ilosci_magazyn)
wart_mal = sum(wart_mag[maska_mal])
print("Wartość produktów z niskim stanem magazynowym: " + str(wart_mal) + "zł.")

suma_mal = sum(ilosci_magazyn[maska_mal])
sum_wszystkich = sum(ilosci_magazyn)

procent = (suma_mal / sum_wszystkich) * 100
print("Stanowią " + str(round(procent)) + "%.")
```

Zadanie 6 (bonus): Analiza korelacji cena–ocena
# Które produkty są "okazją" — ocena >= 4.5 i cena < 1000 zł? Wypisz nazwy i ceny.
# Jaka jest średnia cena produktów z oceną >= 4.0 vs < 4.0?
# Podpowiedź: łączenie warunków: (warunek1) & (warunek2)

```python
maska_okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
prod_okazje = nazwy_a[maska_okazja]
ceny_okazja = ceny[maska_okazja]
print("Produkty które są okazją: " + str(prod_okazje))
print("Ceny: " + str(ceny_okazja))

maska_wysokie = (oceny_klientow >= 4.0)
maska_niskie = (oceny_klientow < 4.0)
srednia_wysokie = np.mean(ceny[maska_wysokie])
srednia_niskie = np.mean(ceny[maska_niskie])
print("Średnia cena produktów z niską oceną: " + str(round(srednia_niskie)))
print("Średnia cena produktów z wysoką oceną: " + str(round(srednia_wysokie)))
```

Ćwiczenie 4: Benchmark — lista vs NumPy (15 min)
Cel
Przekonaj się na własne oczy, jak szybki jest NumPy w porównaniu z listami.

```python
import time
```

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter()- start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter()- start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter()- start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter()- start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter()- start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter()- start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

- NumPy jest **7×** szybszy w sumowaniu-
